```python
import pandas as pd
import numpy as np
import pickle
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("data-process.csv", index_col=[0])
```

Normalization

```python
min_max_scaler = preprocessing.MinMaxScaler()
```

```python
Data = min_max_scaler.fit_transform(df)
```

```python
data2 = pd.DataFrame(Data)
```

```python
for i in range(data2.shape[1]):
    print(data2.loc[:,i].var())
```

```python
data2.describe()
```

Balancing

```python
from imblearn.over_sampling import SMOTE
```

```python
X = data2.iloc[:,:47]
y = data2.iloc[:,47]
```

```python
smote = SMOTE()
X_smote_res, y_smote_res = smote.fit_resample(X,y)
```

Slit dataset

```python
#Separating the training and testing set
X_train, X_test, y_train, y_test = train_test_split(X_smote_res, y_smote_res, train_size=0.8,stratify=y_smote_res)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
with open('data-set.pkl',mode='wb') as f:
    pickle.dump([X_train, y_train, X_test, y_test],f)
```
